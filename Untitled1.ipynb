```python
import pandas as pd
```

```python
df=pd.read_csv("laptop_price.csv",encoding='latin1')
```

```python
import numpy as np
```

```python
df["Ram"]=df["Ram"].str.replace("GB","").astype(int)
```

```python
df["Weight"]=df["Weight"].str.replace("kg","").astype(float)
```

```python
df.head()
```

```python
df["Company"].value_counts()
```

```python
df["Product"].value_counts()
```

```python
df["TypeName"].value_counts()
```

```python
category=["Company","TypeName","ScreenResolution","Cpu","Memory","Gpu","OpSys"]
```

```python
df_new=pd.get_dummies(df, columns=category,drop_first=True)
```

```python
df_new.drop(columns=["laptop_ID","Product"],inplace=True)
```

```python
df_new.head()
```

```python
df_new=df_new.astype(int)
```

```python
df_new
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
exchange_rate_euro_to_inr=110
df_new['Price_inr']=df_new['Price_euros']*exchange_rate_euro_to_inr
```

```python
numerical_features=["Inches","Ram","Weight"]
scaler=StandardScaler()
df_new[numerical_features]=scaler.fit_transform(df_new[numerical_features])
```

```python
df_new.head()
```

```python
X = df_new.drop(columns=["Price_euros","Price_inr"])
Y = df_new["Price_inr"]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.2, random_state=42)
```

```python
X_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()
input_dimension=X_train.shape[1]
print(input_dimension)
```

```python
model.add(Dense(units=128, activation="relu", input_shape=(input_dimension,)))
model.add(Dropout(0.2))

model.add(Dense(units=128, activation="relu"))
model.add(Dropout(0.2))

model.add(Dense(units=64, activation="relu"))
model.add(Dropout(0.2))

model.add(Dense(units=64, activation="relu"))
model.add(Dropout(0.2))

model.add(Dense(units=32, activation="relu"))
model.add(Dropout(0.2))

model.add(Dense(units=32, activation="relu"))
model.add(Dropout(0.2))


model.add(Dense(units=1,activation="linear"))
```

```python
model.summary()
```

```python
early=EarlyStopping(monitor="val_loss",verbose=1,patience=5)
```

```python
model.compile(optimizer="Adam",loss="mean_squared_error",metrics=['mean_absolute_error'])
```

```python
history=model.fit(X_train,y_train,epochs=200,batch_size=30,verbose=1,callbacks=[early],validation_data=(X_test,y_test))
```

```python
y_pred_train=model.predict(X_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("mae:",mae)
print("mse:",mse)
print("r2:",r2)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```
